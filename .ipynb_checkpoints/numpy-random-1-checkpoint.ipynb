```python
from numpy import random
import numpy as np

x = random.randint(100)

print(x) 

# random float

x2 = random.rand()

print(x2) 

# Generate a 1-D array containing 5 random integers from 0 to 100:

x3=random.randint(100, size=(5))

print(x3) 

# Generate a 2-D array with 3 rows, each row containing 5 random integers from 0 to 100:

x4 = random.randint(100, size=(3, 5))

print(x4, np.shape(x4)) 

#  changing the size, therefore the shape 

x5 = random.randint(100, size=(2, 3, 5))

print(x5, np.shape(x5))

# The rand() method also allows you to specify the shape of the array (like in randint)

x6 = random.rand(5)

print(x6) 

# different shape example with random.rand()

x7 = random.rand(3, 5)

print(x7) 

# with choice method we can define an array of values among which we can get one randomly

x8 = random.choice([3, 5, 7, 9])

print(x8) 

# size can also be chosen with choice method 

x9 = random.choice([2,4,6,8], size=(2,3,4))

print(x9)
```

```python
# The choice() method allows us to specify the probability for each value.

# *****The sum of all probability numbers should be 1.*****

# The probability for the value to be 3 is set to be 0.1

# The probability for the value to be 5 is set to be 0.3

# The probability for the value to be 7 is set to be 0.6

# The probability for the value to be 9 is set to be 0

x10 = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(100))

print(x10) 

# same thing with a different shape

x11 = random.choice([3, 5, 7, 9], p=[0.1, 0.3, 0.6, 0.0], size=(3, 5))

print(x11) 
```

```python
# The shuffle() method makes changes to the original array.

arr = np.array([1, 2, 3, 4, 5])

random.shuffle(arr)

print(arr) 

# The permutation() method returns a re-arranged array (and leaves the original array un-changed).

arr2 = np.array([1, 2, 3, 4, 5])

print(random.permutation(arr2)) 
```

```python
import matplotlib.pyplot as plt 
import seaborn as sns 

sns.distplot([0, 1, 2, 3, 4, 5])

plt.show() 
```

```python
sns.distplot([0, 1, 2, 3, 4, 5], hist=False)

plt.show() 
```

```python
# loc - (Mean) where the peak of the bell exists.

# scale - (Standard Deviation) how flat the graph distribution should be.

# size - The shape of the returned array.

x12 = random.normal(size=(2, 3))

print(x12) 

# Generate a random normal distribution of size 2x3 with mean at 1 and standard deviation of 2:

x13 = random.normal(loc=1, scale=2, size=(2, 3))

print(x13) 
```

```python
sns.distplot(random.normal(size=1000), hist=False)

plt.show() 



# Note: The curve of a Normal Distribution is also known as the Bell Curve because of the bell-shaped curve.
```

```python
# It describes the outcome of binary scenarios, e.g. toss of a coin, it will either be head or tails.

# n - number of trials.

# p - probability of occurence of each trial (e.g. for toss of a coin 0.5 each).

# size - The shape of the returned array.

x13 = random.binomial(n=10, p=0.5, size=10)
x14 = random.binomial(n=10, p=0.9, size=10)

print(x13) 
print(x14) 
```

```python
sns.distplot(random.binomial(n=10, p=0.5, size=1000), hist=True, kde=False)

plt.show()
```

```python
# normal distribution is continous whereas binomial is discrete,but if there are enough 
# data points it will be quite similar to normal distribution with certain loc and scale.

sns.distplot(random.normal(loc=50, scale=5, size=1000), hist=False, label='normal')
sns.distplot(random.binomial(n=100, p=0.5, size=1000), hist=False, label='binomial')

plt.show() 
```

```python
# Poisson Distribution is a Discrete Distribution.

# It estimates how many times an event can happen in a specified time.
# e.g. If someone eats twice a day what is probability he will eat thrice?

# lam - rate or known number of occurences e.g. 2 for above problem.

# size - The shape of the returned array.

# Generate a random 1x10 distribution for occurence 2

x15 = random.poisson(lam=2, size=10)

print(x15) 
```

```python
sns.distplot(random.poisson(lam=2, size=1000), kde=False)

plt.show() 
```

```python
# we can see that similar to binomial for a large enough poisson distribution 
# it will become similar to normal distribution with certain std dev and mean.

sns.distplot(random.normal(loc=50, scale=7, size=1000), hist=False, label='normal')
sns.distplot(random.poisson(lam=50, size=1000), hist=False, label='poisson')

plt.show() 
```

```python
# binomial distribution is for discrete trials, whereas poisson distribution is for continuous trials.
# But for very large n and near-zero p binomial distribution is near identical to 
# poisson distribution such that n * p is nearly equal to lam.

sns.distplot(random.binomial(n=1000, p=0.01, size=1000), hist=False, label='binomial')
sns.distplot(random.poisson(lam=10, size=1000), hist=False, label='poisson')

plt.show() 
```

```python
# Uniform distribution is used to describe probability where every event has equal chances of occuring.

# E.g. Generation of random numbers.

# a - lower bound - default 0 .0

# b - upper bound - default 1.0

# size - The shape of the returned array.

x16 = random.uniform(size=(2, 3))

print(x16) 
```

```python
sns.distplot(random.uniform(size=1000), hist=False)

plt.show() 
```

```python
# Logistic Distribution is used to describe growth.

# loc - mean, where the peak is. Default 0.

# scale - standard deviation, the flatness of distribution. Default 1.

# size - The shape of the returned array.

x17 = random.logistic(loc=1, scale=2, size=(2, 3))

print(x17) 
```

```python
sns.distplot(random.logistic(size=1000), hist=False)

plt.show() 
```

```python
# For higher value of scale (standard deviation) the normal and 
# logistic distributions are near identical apart from the peak.

sns.distplot(random.normal(scale=2, size=1000), hist=False, label='normal')
sns.distplot(random.logistic(size=1000), hist=False, label='logistic')

plt.show() 
```

```python
# Multinomial distribution is a generalization of binomial distribution.

# It describes outcomes of multi-nomial scenarios. e.g. Blood type of a population, dice roll outcome.

# n - number of possible outcomes (e.g. 6 for dice roll).

# pvals - list of probabilties of outcomes (e.g. [1/6, 1/6, 1/6, 1/6, 1/6, 1/6] for dice roll).

# size - The shape of the returned array.

x18 = random.multinomial(n=6, pvals=[1/6, 4/6, 0/6, 0/6, 0/6, 1/6])

x19 = random.multinomial(n=6, pvals=[1/6, 1/6, 1/6, 1/6, 1/6, 1/6])

x20 = random.multinomial(n=6, pvals=[0/6, 0/6, 0/6, 1/6, 2/6, 3/6])

print(x18, "<== x18") 
print(x19, "<== x19")
print(x20, "<== x20") 

# ***** Multinomial samples will NOT produce a single value! They will produce one value for each pval. *****
```

```python
# Exponential distribution is used for describing time till next event e.g. failure/success etc.

# scale - inverse of rate ( see lam in poisson distribution (which was rate or number of occurence) ) defaults to 1.0.

# size - The shape of the returned array.


x21 = random.exponential(scale=2, size=(2, 3))

print(x21) 
```

```python
sns.distplot(random.exponential(size=1000), hist=False)

plt.show() 
```

```python
# Chi Square distribution is used as a basis to verify the hypothesis

# df - (degree of freedom)

# size - The shape of the returned array

x22 = random.chisquare(df=2, size=(2, 3))

print(x22) 
```

```python
sns.distplot(random.chisquare(df=1, size=1000), hist=False)

plt.show() 
```

```python
# Rayleigh distribution is used in signal processing.

# scale - (standard deviation) decides how flat the distribution will be default 1.0).

# size - The shape of the returned array.

x23 = random.rayleigh(scale=2, size=(2, 3))

print(x23) 
```

```python
sns.distplot(random.rayleigh(size=1000), hist=False)

plt.show() 
```

```python
# A distribution following Pareto's law i.e. 80-20 distribution (20% factors cause 80% outcome).

# a - shape parameter.

# size - The shape of the returned array.

x24 = random.pareto(a=2, size=(2, 3))

print(x24) 
```

```python
sns.distplot(random.pareto(a=2, size=1000), kde=False)

plt.show() 
```

```python
# Zipf's Law: In a collection the nth common term is 1/n times of the most common term.
# E.g. 5th common word in english has occurs nearly 1/5th times as of the most used word.

# a - distribution parameter

# size - The shape of the returned array

x25 = random.zipf(a=2, size=(2, 3))

print(x25) 
```

```python
# Sample 1000 points but plotting only ones with value < 10 for more meaningful chart.

x26 = random.zipf(a=2, size=1000)
sns.distplot(x26[x26<10], kde=False)
```
